````python
# BMR laki - laki : BMR = 66 + (13,7*BB) + (5*TB) + (6,8*U)



bb = int(input('Masukan Berat Badan anda (kg) : '))
tb = int(input('Masukan Tinggi Badan anda (cm) : '))
u = int(input('Masukan usia anda (tahun) : '))

def bmr_male (bb,tb,u) :
  bmr_mcalc = int(66 + (13.7*bb) + (5*tb) + (6.8*u))
  if bmr_mcalc > 0 :
    print(f'BMR anda adalah : {bmr_mcalc}')
  else :
    print(f'Angka tidak valid')

print(bmr_male(bb,tb,u))
````

````python
# BMR Perempuan : BMR = 655 + (9,6*BB) + (1,8*TB) + (4,7*U)
#BB adalah Berat Badan dalam satuan Kilogram
#TB adalah Tinggi Badan dalam satuan sentimeter
#U adalah Usia dalam satuan tahun

bb = int(input('Masukan Berat Badan anda (kg) : '))
tb = int(input('Masukan Tinggi Badan anda (cm) : '))
u = int(input('Masukan usia anda (tahun) : '))

def bmr_female (bb,tb,u) :
  bmr_fcalc = int(655 + (9.6*bb) + (1.8*tb) + (4.7*u))
  if bmr_fcalc > 0 :
    print(f'BMR anda adalah : {bmr_fcalc}')
  else :
    print(f'Angka tidak valid')

print(bmr_female(bb,tb,u))
````
